```python
##The task is to come up with a optimization software that allows the battery to trade in 3 wholesale markets across the period
##of 1 year

##We will be using pyomo for this undertaking, modeling the problem as an mixed-integer linear program (MILP)
```

```python
##Importing the relevant packages
import os 
root_dir = os.path.abspath(os.path.join(os.getcwd(), '..', '')) + '//'
data_dir = root_dir + 'initial_materials//'
save_dir = root_dir + 'results//'
solver_path = root_dir + 'ampl//'
##Adding so that the solver can use it 
os.environ['PATH'] = os.environ['PATH'] + ';' + solver_path
os.environ['LD_LIBRARY_PATH'] = solver_path
import pandas as pd 
import pyomo.environ as pyo 
from pyomo.opt import SolverStatus
import matplotlib.pyplot as plt
from joblib import Parallel, delayed
```

```python
##Defining the parameters of the battery 
batt_param_dict = {'charging': {'rate': {'value': 2, 
                                         'unit': 'MW'},
                                'efficiency': 0.05},
                   'discharging': {'rate': {'value': 2, 
                                            'unit': 'MW'},
                                   'efficiency': 0.05},
                   'storage': {'volume': {'value': 4,
                                          'unit': 'MWh'},
                               'degredation': {'value': 0.001,
                                               'unit': '%/cycle'}},
                   'life': {'time': {'value': 10, 'unit': 'year'},
                            'cycles': {'value': 5000, 'unit': 'cycle'}},
                   'cost': {'capital': {'value': 500000, 'unit': '£'},
                            'operational': {'value': 5000, 'unit': '£'}}} 
```

```python
##Defining the market parameters 
market_param_dict = [{'name': 'market_1',
                      'trading_interval': {'value': 30, 
                                           'unit': 'min'}},
                     {'name': 'market_2',
                      'trading_interval': {'value': 30, 
                                           'unit': 'min'}},
                     {'name': 'market_3',
                      'trading_interval': {'value': 1440, 
                                           'unit': 'min'}}]
```

```python
##Loading the data 
price_data_1 = pd.read_excel(data_dir + 'Second Round Technical Question - Attachment 2.xlsx', sheet_name = 'Half-hourly data')
price_data_1['timestamp'] = pd.to_datetime(price_data_1['Unnamed: 0'])
del price_data_1['Unnamed: 0']
price_data_1.set_index(['timestamp'], inplace = True)
price_data_1_last_time_stamp = price_data_1.index[-1]
price_data_1['supply'] = sum(price_data_1[x] for x in ['Wind Generation [MW]', 'Solar Generation [MW]', 'Coal Generation [MW]',
                                                       'Gas Generation [MW]'])
price_data_1['short_fall'] = price_data_1['supply'] - price_data_1['Transmission System Electricity Demand [MW]']

price_data_2 = pd.read_excel(data_dir + 'Second Round Technical Question - Attachment 2.xlsx', sheet_name = 'Daily data')
price_data_2['timestamp'] = pd.to_datetime(price_data_2['Unnamed: 0'])
del price_data_2['Unnamed: 0']
last_value = price_data_2['Market 3 Price [£/MWh]'].iloc[-1]
price_data_2 = pd.concat([price_data_2, pd.DataFrame(data = [[last_value, price_data_1_last_time_stamp]], 
                                                     columns = ['Market 3 Price [£/MWh]', 'timestamp'])])
price_data_2.set_index(['timestamp'], inplace = True, drop = True)
price_data_2 = price_data_2.resample('30min').ffill()
```

```python
##Creating the demand and market prices 
    ##Curating and unit management (£/MW)
opti_data = pd.DataFrame()
opti_data['timestamp'] = price_data_1.index
opti_data['market_1'] = [x / 2 for x in price_data_1['Market 1 Price [£/MWh]']]
opti_data['market_2'] = [x / 2 for x in price_data_1['Market 2 Price [£/MWh]']]
opti_data['market_3'] = [x / 2 for x in price_data_2['Market 3 Price [£/MWh]']]
opti_data['demand'] = [x for x in price_data_1['Transmission System Electricity Demand [MW]']]
opti_data['supply'] = [x for x in price_data_1['supply']]
opti_data['short_fall'] = [x for x in price_data_1['short_fall']]
```

```python
##Base on the plots, we can comfortably fall into the price taker's assumption --- the battery's capacity is very small compared
##to the magnitude of the shortfall quantums
```

```python
##Here is what we are going to do:
    ##1. Assume that we are looking back in time, i.e., perfect price information/ have a perfect price forecast model
    ##2. Create a software which solves the problem for a defined period of time
    ##3. Within that period of time make assumptions on the following:
        ##a. degradation to the storage volume 0.001/100 ~= 0 (it is very small, deal with it in the future, not in this test)
        ##b. fix the number of permissible cycles in the given period of time
        ##c. we must keep the battery operational, regardless of whether we trade or not
```

```python
##Let the period of time be defined as a day 
opti_data['date'] = [x.date() for x in opti_data['timestamp']]
```

```python
##Executing the program 
unique_dates = list(opti_data['date'].unique())
```

```python
##Formulating the optimization problem 
def create_opti_model (uniq_date, batt_param_dict, market_param_dict, opti_data):
    ###########################
    ##Grab the necessary data 
    ###########################
    curr_data = opti_data[opti_data['date'] == uniq_date].copy()
    curr_data.reset_index(inplace = True, drop = True)
    
    ###########################
    ##Initializing the model 
    ###########################
    model = pyo.ConcreteModel()
    
    ###########################
    ##Adding the sets 
    ###########################
    num_markets = int(len(market_param_dict))
    
    model.set_markets = pyo.Set(initialize = range(1, num_markets + 1))
    model.block_market = pyo.Block(list(range(1, num_markets + 1)))
    for market_idx in model.set_markets:
        model.block_market[market_idx].set_time_blocks = pyo.Set(initialize = range(1, int(curr_data.shape[0]) + 1))
        
    model.set_time_blocks = pyo.Set(initialize = range(1, int(curr_data.shape[0]) + 1))
    
    ###########################
    ##Adding the parameters 
    ###########################
        ##The cycle limit (assume a day)
    cycle_limit = batt_param_dict['life']['cycles']['value'] / (365 * batt_param_dict['life']['time']['value'])
    model.param_cycle_limit = pyo.Param(name = 'cycle_limit', domain = pyo.Reals, default = cycle_limit)
        ##The charging rate 
    model.param_charging_rate = pyo.Param(name = 'charging_rate', domain = pyo.Reals, 
                                          default = batt_param_dict['charging']['rate']['value']) 
        ##The charging efficiency 
    model.param_charging_efficiency = pyo.Param(name = 'charging_efficiency', domain = pyo.Reals, 
                                                default = batt_param_dict['charging']['efficiency'])
        ##The discharging rate 
    model.param_discharging_rate = pyo.Param(name = 'discharging_rate', domain = pyo.Reals, 
                                             default = batt_param_dict['discharging']['rate']['value'])
        ##The discharging efficiency 
    model.param_discharging_efficiency = pyo.Param(name = 'discharging_efficiency', domain = pyo.Reals, 
                                                   default = batt_param_dict['discharging']['efficiency'])
        ##The storage volume 
    model.param_storage_volume = pyo.Param(name = 'storage_volume', domain = pyo.Reals, 
                                           default = batt_param_dict['storage']['volume']['value'])
        ##The initial volume 
    model.param_initial_volume = pyo.Param(name = 'initial_volume', domain = pyo.Reals, default = 0)
        
        ##The market data 
    for market_idx in model.set_markets:
            ##The market price
        model.block_market[market_idx].param_price = pyo.Param(model.block_market[market_idx].set_time_blocks, 
                                                               name = 'market_price', domain = pyo.Reals, mutable = True)
        for time_block in model.block_market[market_idx].set_time_blocks:
            model.block_market[market_idx].param_price[time_block] = curr_data['market_' + str(market_idx)][time_block - 1]
    
    ###########################
    ##Adding the variables 
    ###########################
        ##The battery's charging quantity 
    model.var_charge = pyo.Var(model.set_time_blocks, name = 'charge value per time-block', domain = pyo.Reals,
                               bounds = (0, None))
        ##The battery's discharging quantity 
    model.var_discharge = pyo.Var(model.set_time_blocks, name = 'discharge value per time-block', domain = pyo.Reals,
                                 bounds = (None, 0))  
        ##A binary variable to determine if it is in charging or discharging mode
    model.var_is_charge = pyo.Var(model.set_time_blocks, name = 'differentiate between charging or discharging', domain = pyo.Binary)
    
        ##The market-level variables 
    for market_idx in model.set_markets:
        model.block_market[market_idx].var_buy_sell = pyo.Var(model.block_market[market_idx].set_time_blocks, 
                                                              name = 'buying or selling', domain = pyo.Reals, 
                                                              bounds = (-4, 4))

    ###########################
    ##Adding the expressions
    ###########################
    ##The efficiency factored charge per time-block 
    def expr_efficiency_factored_charge (model, time_block):
        expr = model.var_charge[time_block] * (1 - model.param_charging_efficiency)
        
        return expr
    model.expr_efficiency_factored_charge = pyo.Expression(model.set_time_blocks, rule = expr_efficiency_factored_charge)
    
    ##The efficiency factored discharge per time-block 
    def expr_efficiency_factored_discharge (model, time_block):
        expr = model.var_discharge[time_block] * 1 / (1 - model.param_charging_efficiency)
        
        return expr
    model.expr_efficiency_factored_discharge = pyo.Expression(model.set_time_blocks, rule = expr_efficiency_factored_discharge)
    
    ##The stored volume at the start of the time-block 
    def expr_stored_volume_start_time_block (model , time_block):
        expr = 0 
        if time_block == 1:
            expr = model.param_initial_volume
        else:
            ##Factor in efficiency and also MW - MWh
            expr = model.expr_stored_volume_at_start[time_block - 1] 
            expr = expr + (model.expr_efficiency_factored_charge[time_block - 1] / 2)
            expr = expr + (model.expr_efficiency_factored_discharge[time_block - 1] / 2)
        
        return expr
    model.expr_stored_volume_at_start = pyo.Expression(model.set_time_blocks, rule = expr_stored_volume_start_time_block)
    
    ##The market-level nett volume transacted 
    def expr_nett_market_volume (model, time_block):
        expr = 0 
        for market_idx in model.set_markets:
            expr = expr + model.block_market[market_idx].var_buy_sell[time_block]
        return expr
    model.expr_market_nett_volume_transacted = pyo.Expression(model.set_time_blocks, rule = expr_nett_market_volume)
    
    ##The market-level revenue 
    def expr_market_revenue (market_model, time_block):
        expr = market_model.param_price[time_block] * -market_model.var_buy_sell[time_block]
        return expr
    
    for market_idx in model.set_markets:
        model.block_market[market_idx].expr_revenue = pyo.Expression(model.block_market[market_idx].set_time_blocks,
                                                                     rule = expr_market_revenue)
        
    ##The total revenue per time-block
    def expr_total_revenue_per_time_block (model, time_block):
        expr = 0 
        for market_idx in model.set_markets:
            expr = expr + model.block_market[market_idx].expr_revenue[time_block]
        return expr 
    
    model.expr_total_revenue_per_time_block = pyo.Expression(model.set_time_blocks, rule = expr_total_revenue_per_time_block)
    
    ##The total revenue 
    def expr_total_revenue (model):
        return sum(model.expr_total_revenue_per_time_block[time_block] for time_block in model.set_time_blocks)
    
    model.expr_total_revenue = pyo.Expression(rule = expr_total_revenue)
    
    ###########################
    ##Adding the constraints 
    ###########################    
    ##The storage limits cannot be violated 
    model.const_storage_limit = pyo.ConstraintList()
    for time_block in model.set_time_blocks:
        ##The lower bound 
        expr = model.expr_stored_volume_at_start[time_block] >= 0 
        model.const_storage_limit.add(expr)
        ##The upper bound 
        expr = model.expr_stored_volume_at_start[time_block] <= model.param_storage_volume
        model.const_storage_limit.add(expr)
        
    ##The charging and discharging limits
    model.const_charging_discharging_limits = pyo.ConstraintList()
    for time_block in model.set_time_blocks:
        ##Charging
            ##Lower bound 
        expr = model.var_charge[time_block] >= 0 
        model.const_charging_discharging_limits.add(expr)
            ##Upper bound 
        expr = model.var_charge[time_block] <= model.param_charging_rate
        model.const_charging_discharging_limits.add(expr)
        ##Discharging 
            ##Lower bound 
        expr = model.var_discharge[time_block] <= 0 
        model.const_charging_discharging_limits.add(expr)
            ##Upper bound 
        expr = model.var_discharge[time_block] >= -model.param_discharging_rate
        model.const_charging_discharging_limits.add(expr)
        
    ##Connecting the charge and discharge variables with the nett volume transacted 
    model.const_charging_discharging_connector = pyo.ConstraintList()
    
    BIG_M = 1000
    for time_block in model.set_time_blocks:
        ##Charging 
            ##Lower bound 
        expr = model.expr_market_nett_volume_transacted[time_block] - (BIG_M * (1 - model.var_is_charge[time_block]))
        expr = expr <= model.var_charge[time_block]
        model.const_charging_discharging_connector.add(expr)
            ##Upper bound 
        expr = model.expr_market_nett_volume_transacted[time_block] + (BIG_M * (1 - model.var_is_charge[time_block]))
        expr = expr >= model.var_charge[time_block]
        model.const_charging_discharging_connector.add(expr)
        ##Discharging 
            ##Lower bound 
        expr = model.expr_market_nett_volume_transacted[time_block] - (BIG_M * model.var_is_charge[time_block])
        expr = expr <= model.var_discharge[time_block]
        model.const_charging_discharging_connector.add(expr)
            ##Upper bound 
        expr = model.expr_market_nett_volume_transacted[time_block] + (BIG_M * model.var_is_charge[time_block])
        expr = expr >= model.var_discharge[time_block]
        model.const_charging_discharging_connector.add(expr)
    
    ##To ensure that charging and discharging values is always connected to the binary variable 
    model.const_charging_discharging_binary_connector = pyo.ConstraintList()
    for time_block in model.set_time_blocks:
        ##Charging 
        expr = model.var_charge[time_block] <= model.var_is_charge[time_block] * model.param_charging_rate
        model.const_charging_discharging_binary_connector.add(expr)
        ##Discharging 
        expr = model.var_discharge[time_block] >= (1 - model.var_is_charge[time_block]) * -model.param_discharging_rate
        model.const_charging_discharging_binary_connector.add(expr)
    
    ##To ensure that the cycles consumed within the given day is respected 
    model.const_cycles_consumed = pyo.ConstraintList()
    
    expr_charge = 0 
    expr_discharge = 0 
    for time_block in model.set_time_blocks:
        expr_charge = expr_charge + model.var_charge[time_block]
        expr_discharge = expr_discharge + model.var_discharge[time_block]
    
    expr_charge = expr_charge <= model.param_cycle_limit * model.param_storage_volume
    model.const_cycles_consumed.add(expr_charge)
    
    expr_discharge = expr_discharge >= -model.param_cycle_limit * model.param_storage_volume
    model.const_cycles_consumed.add(expr_discharge)
    
    ##The equality constraint depending on the market 
    market_idx = 3
    model.block_market[market_idx].const_multiple_time_block_equality = pyo.ConstraintList()
    
    for time_block in model.block_market[market_idx].set_time_blocks:
        if time_block > 1:
            expr = model.block_market[market_idx].var_buy_sell[time_block] == model.block_market[market_idx].var_buy_sell[time_block - 1]
            model.block_market[market_idx].const_multiple_time_block_equality.add(expr)

    ###########################
    ##Adding the objective function 
    ########################### 
    model.objective_function = pyo.Objective(expr = model.expr_total_revenue, sense = pyo.maximize)
    
    ###########################
    ##Solving the problem  
    ########################### 
    solver = 'highs.exe'
    executable_path = solver_path + solver 
    tee = False 
    options = {'timelimit': 60}
    
    opt = pyo.SolverFactory(solver, executables = executable_path, solver_io = 'nl')
    model.solver_output = opt.solve(model, tee = tee, options = options)
    
    ###########################
    ##Packaging the result 
    temp_data = []
    for time_block in model.set_time_blocks:
        curr_temp_data = [curr_data['timestamp'][time_block - 1]]
        if model.solver_output.solver.status == pyo.SolverStatus.ok:
            for market_idx in model.set_markets:
                curr_temp_data.append(model.block_market[market_idx].var_buy_sell[time_block].value)
                curr_temp_data.append(model.block_market[market_idx].expr_revenue[time_block].expr())
            
            curr_temp_data.append(model.var_charge[time_block].value)
            curr_temp_data.append(model.var_discharge[time_block].value)
            curr_temp_data.append(model.expr_stored_volume_at_start[time_block].expr())
            curr_temp_data.append(model.expr_total_revenue_per_time_block[time_block].expr())
            
            temp_data.append(curr_temp_data)
        else:
            for market_idx in model.set_markets:
                curr_temp_data.append(0)
                curr_temp_data.append(0)
            
            curr_temp_data.append(0)
            curr_temp_data.append(0)
            curr_temp_data.append(model.expr_stored_volume_at_start[time_block].expr())
            curr_temp_data.append(0)
                
            temp_data.append(curr_temp_data)

    ret_df_columns = ['timestamp']
    for market_idx in model.set_markets:
        ret_df_columns.append('market_' + str(market_idx) + '_quantity')
        ret_df_columns.append('market_' + str(market_idx) + '_revenue')
    
    ret_df_columns = ret_df_columns + ['charge', 'discharge', 'stored', 'revenue']
    
    ret_df = pd.DataFrame(data = temp_data, columns = ret_df_columns)
    
    model.return_results = ret_df
    
    return model.return_results
```

```python
##Do a parallel solve
ret_df_list = Parallel(n_jobs = 32)(delayed(create_opti_model)(i, batt_param_dict, market_param_dict, opti_data) for i in unique_dates)
```

```python
##Saving the results 
ret_df = pd.DataFrame()

for sub_df in ret_df_list:
    ret_df = pd.concat([ret_df, sub_df], ignore_index = True)

ret_df.to_csv(save_dir + 'results.csv')
```

```python
##Checking the profit 
ret_df['year'] = [x.year for x in ret_df['timestamp']]
unique_years = list(ret_df['year'].unique())

count = 0
carry_over = 0 
temp_data = []
for year in unique_years:
    curr_df = ret_df[ret_df['year'] == year].copy()
    revenue = sum(curr_df['revenue'])
    cost = batt_param_dict['cost']['operational']['value']
    if count == 0:
        cost = cost + batt_param_dict['cost']['capital']['value'] 
    if carry_over < 0:
        cost = cost - carry_over
    profit = revenue - cost
    if profit < 0:
        carry_over = profit
    else:
        carry_over = 0
    count = count + 1
    
    temp_data.append([year, revenue, profit])
    print('Year: ' + str(year) + ', Revenue: ' + str(revenue) + ', Profit: ' + str(profit))

summary_df = pd.DataFrame(data = temp_data, columns = ['year', 'revenue', 'profit'])
summary_df.to_csv(save_dir + 'summary.csv')
```
